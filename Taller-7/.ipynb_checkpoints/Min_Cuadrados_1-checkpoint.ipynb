```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F=(lambda x: 2*x-2,\
   lambda x:(1-x)/2,\
   lambda x: 4-x )
```

```python
x=np.linspace(-5,5,300)
y=np.linspace(-5,5,300)
```

```python
A=np.array([[2,-1],[1,2],[1,1]])
B=np.array([[2,1,1],[-1,2,1]])
v = np.array([[2],[1],[4]])
```

```python
Bv= np.dot(B,v)
BA= np.dot(B,A)
```

```python
sol= (np.linalg.solve(BA,Bv))
xsol=float(sol[0])
ysol=float(sol[0])
print('El punto solución es:\n(%.3f,%.3f)'%(xsol,ysol))
```
