```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=int(1e4)
```

```python
def esfera(N,dim=3):
    vector = np.random.randn(dim, N)
    vector /= np.linalg.norm(vector, axis=0)
    array=np.random.uniform(size=N)
    return vector*array
```

```python
xi, yi, zi = esfera(N)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection = '3d')
#configurar lmites de los ejes
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)
ax.view_init(20, 20)
ax.set_box_aspect([1,1,1])
ax.scatter(xi, yi, zi, s=1, c="r",)
```
